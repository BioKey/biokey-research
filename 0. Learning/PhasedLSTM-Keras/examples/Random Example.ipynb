```python
import numpy as np
from keras.layers import LSTM
from keras.models import Sequential, load_model
from keras.utils import np_utils

from phased_lstm_keras.PhasedLSTM import PhasedLSTM as PLSTM
```

```python
X = np.random.random((32, 100, 2))
Y = np.random.random((32, 1))
```

```python
model_lstm = Sequential()
model_lstm.add(LSTM(1, input_shape=(100, 2)))
model_lstm.compile('rmsprop', 'mse', metrics=['accuracy'])
model_lstm.save('model_lstm.h5')
model_lstm = load_model('model_lstm.h5')
model_lstm.summary()
```

```python
model_lstm.fit(X, Y, epochs=100)
```

```python
model_lstm.predict(X[:3])
```

```python
model_plstm = Sequential()
model_plstm.add(PLSTM(1, input_shape=(100, 2), implementation=2))
model_plstm.compile('rmsprop', 'mse', metrics=['accuracy'])
# model_plstm.save('model_plstm.h5')
# model_plstm = load_model('model_plstm.h5')
model_plstm.summary()
```
