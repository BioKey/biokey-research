```python
data = BiokeyData(credentials.postgres)
```

```python
data.get_dwells().groupby('user_id').dwell
```

```python
dwells = data.get_dwells("1d63b44d-a7cb-4ee6-b228-b0ff5b7d086f")
```

```python
dwells['Interval'] = (dwells.down - dwells.down.shift(1))
dwells['Sequence'] = ((dwells.Interval < dwells.Interval.quantile(0.95)) != (dwells.Interval < dwells.Interval.quantile(0.99)).shift(1)).cumsum()
```

```python
sequence_long_enough = dwells.Sequence.value_counts() > 10
dwells_in_sequences = dwells.loc[dwells.Sequence.apply(lambda x: sequence_long_enough[x])]
```

```python
seq_prob = (dwells_in_sequences.Sequence.value_counts()/dwells_in_sequences.Sequence.value_counts().sum())
dwells_in_sequences['SeqProb'] = dwells_in_sequences.Sequence.apply(lambda x: seq_prob[x])
```

```python
unique_seq = dwells_in_sequences.Sequence.unique()
unique_seq_prob = np.apply_along_axis(lambda x: seq_prob[x], axis=0, arr=unique_seq)
```

```python
sample_rate = 0.5
num_samples = int(np.rint(len(unique_seq)*sample_rate))
```

```python
seq_to_include = np.random.choice(unique_seq, size=num_samples, replace=False, p=unique_seq_prob)
dwells.loc[dwells.Sequence.isin(seq_to_include)]
```
